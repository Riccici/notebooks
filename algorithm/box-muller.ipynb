Box-Muller 生成正态分布随机数

```python
import math
import random
import matplotlib.pyplot as plt

PI = math.pi

def log_func():
    x = random.random()
    return math.log(x)

def z0_func():
    u1 = random.random()  # Get a random float in the range [0, 1)
    u2 = random.random()
    return math.sqrt(-2.0 * math.log(u1)) * math.cos(2.0 * PI * u2)

def get_normal_random(mean, standard_deviation):
    u1 = random.random()  # Get a random float in the range [0, 1)
    u2 = random.random()

    if u1 <= 0.0:
        u1 = 0.0001
    if u2 <= 0.0:
        u2 = 0.0001

    z0 = math.sqrt(-2.0 * math.log(u1)) * math.cos(2.0 * PI * u2)
    # z1 = math.sqrt(-2.0 * math.log(u1)) * math.sin(2.0 * PI * u2)  # If you need a second independent normal random number

    return mean + z0 * standard_deviation
# Generate 10000 random numbers
sample_num = 100000

log_samples = [log_func() for _ in range(sample_num)]
z0_samples = [z0_func() for _ in range(sample_num)]
samples = [get_normal_random(0, 0.5) for _ in range(sample_num)]

# Plot histogram
plt.hist(log_samples, bins=50, density=True, alpha=0.6, color='g')
plt.title('Distribution of log_func')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Plot histogram
plt.hist(z0_samples, bins=50, density=True, alpha=0.6, color='g')
plt.title('Distribution of log_func')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Plot histogram
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g')
plt.title('Distribution of get_normal_random')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
